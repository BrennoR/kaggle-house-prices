## Data Preprocessing

```python
train_set = data.drop('SalePrice', axis=1)
train_labels = data['SalePrice'].copy()
```

```python
train_null_values = train_set[train_set.columns[train_set.isnull().any(axis=0)]].isnull().sum()
train_null_values.sort_values(ascending=False)
```

```python
train_set_num = train_set.select_dtypes(include=[np.number])
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="median")
```

```python
imputer.fit(train_set_num)
```

```python
imputer.statistics_
```

```python
X_num = imputer.transform(train_set_num)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minMax = MinMaxScaler()
minMax.fit(X_num)
X_num = minMax.transform(X_num)
```

```python
X_num = pd.DataFrame(X_num, columns=train_set_num.columns)
```

```python
num_cols = train_set.select_dtypes(include=[np.number]).columns
cat_cols = train_set.select_dtypes(include=[object]).columns

train_cat = train_set[cat_cols]

train_cat_null_values = train_cat[train_cat.columns[train_cat.isnull().any(axis=0)]].isnull().sum()
train_cat_null_values.sort_values(ascending=False)
```

```python
X_cat = train_cat.fillna('None')
```

```python
X = pd.concat([X_num, X_cat], axis=1)
X.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_attribs = list(X_cat)

oneHot = ColumnTransformer([
        ("cat", OneHotEncoder(), cat_attribs),
    ])

X_prepared = oneHot.fit_transform(X)
```
