## Data Preprocessing

```python
train_set = data.drop('SalePrice', axis=1)
train_labels = data['SalePrice'].copy()
```

```python
train_null_values = train_set[train_set.columns[train_set.isnull().any(axis=0)]].isnull().sum()
train_null_values.sort_values(ascending=False)
```

```python
train_set_num = train_set.select_dtypes(include=[np.number])
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="median")
```

```python
imputer.fit(train_set_num)
```

```python
imputer.statistics_
```

```python
X_num = imputer.transform(train_set_num)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minMax = MinMaxScaler()
minMax.fit(X_num)
X_num = minMax.transform(X_num)
```

```python
X_num = pd.DataFrame(X_num, columns=train_set_num.columns)
```

```python
num_cols = train_set.select_dtypes(include=[np.number]).columns
cat_cols = train_set.select_dtypes(include=[object]).columns

train_cat = train_set[cat_cols]

train_cat_null_values = train_cat[train_cat.columns[train_cat.isnull().any(axis=0)]].isnull().sum()
train_cat_null_values.sort_values(ascending=False)
```

```python
X_cat = train_cat.fillna('None')
```

```python
X = pd.concat([X_num, X_cat], axis=1)
X.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_attribs = list(X_cat)

oneHot = ColumnTransformer([
        ("cat", OneHotEncoder(), cat_attribs),
    ])

X_prepared = oneHot.fit_transform(X)
```

## Model Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lin_reg = LinearRegression()

lin_reg_scores = cross_val_score(lin_reg, X_prepared, train_labels,
                         scoring="neg_mean_squared_log_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_reg_scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(lin_rmse_scores)
```

```python
def score_model(regressor):
    scores = cross_val_score(regressor, X_prepared, train_labels,
                            scoring="neg_mean_squared_log_error", cv=10)
    rmse_score = np.sqrt(-scores)
    display_scores(rmse_score)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
score_model(ridge)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
score_model(lasso)
```

```python
from sklearn.linear_model import ElasticNet

elastic_net = ElasticNet()
score_model(elastic_net)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor()
score_model(knn)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor()
score_model(tree)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
score_model(forest)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
score_model(xgb)
```

## Fine Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'eta': [0.1, 0.3, 0.5], 'gamma': [0, 1, 10, 100], 'max_depth': [0, 2, 4, 6, 8, 10],
    'alpha': [0, 1, 10]}
  ]

xgb_final = XGBRegressor()
grid_search = GridSearchCV(xgb_final, param_grid, cv=5,
                           scoring='neg_mean_squared_log_error', return_train_score=True)
grid_search.fit(X_prepared, train_labels)
```
